```python
columns = {
        'Company':'_y', 
        'Brief Description':'_y',
        'Founders':'_y', 
        'Company Focus':'_x', 
        'Founded':'_x', 
        'Protein Category':'_y',
        'Ingredient Type':'_x',
        'Technology Focus':'_x', 
        'Animal-Type Analog':'_x',
        'Country':'_y', 
        'City':'_y'}

def columns_correction(df, columns):
    
    def condition_label(value, condition):
        
        if value == '_y' and condition == 1:
            return 'right_only'
        elif value == '_y' and condition == 2:
            return 'left_only'
        elif value == '_x' and condition == 1:
            return 'left_only'
        elif value == '_x' and condition == 2:
            return 'right_only'

    labels = {'_y':'right_only', '_x':'left_only'}
    def opposite(value):
        if value == '_y':
            return '_x'
        elif value == '_x':
            return '_y'

    

# Company Y, BRIEF DESCRIPTION Y, Founders y, Company Focus_x, Founded_x, Protein Category_y, Ingredient Type_x,
# Technology Focus_x, Animal-Type Analog_x, Country_y, City_y

    exclude = []
    for i in columns:
        c1 = (((df['_merge']=='both') & (df[i+columns[i]].isnull()==False)) | (df['_merge']==labels[columns[i]]))
        c2 = (((df['_merge']=='both') & (df[i+columns[i]].isnull()==True)) | (df['_merge']==labels[opposite(columns[i])]))
        r1 = df[i+columns[i]]
        r2 = df[i+opposite(columns[i])]
        conditions = [c1,c2]
        results = [r1,r2]
        df[i] = np.select(conditions,results)

        exclude.append(i+'_x')
        exclude.append(i+'_y')
    exclude.append('_merge')
    df = df.drop(columns=exclude)
    
    return df

```

```python
all_companies = columns_correction(all_companies_2, columns)
all_companies = all_companies[['Company', 'Brief Description', 'Founded', 'Founders',
 'Company Focus', 'Primary focus', 'Business model', 'Protein Category',
 'Ingredient Type', 'Product Type', 'Technology Focus', 'Animal-Type Analog',
 'Website', 'Operating Regions', 'City', 'Country']]
```
